# Importar las librerías

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importamos de la librería sklearn, la función train_test_split
from sklearn.model_selection import train_test_split
# Importamos la función StandardScaler para normalizar las variables
from sklearn.preprocessing import StandardScaler
# Importamos la función LinearRegression de la librería sklearn para inicializar el modelo
from sklearn.linear_model import LinearRegression
```

# Importar el dataset

Este conjunto de datos tiene datos recopilados de Nueva York, California y Florida sobre 50 startups, 17 en cada estado. Las variables utilizadas en el conjunto de datos son  el beneficio, Gasto en I+D, Gasto en administración y Gasto en marketing. La variable objetivo a predecir en este caso va a ser el beneficio ("Profit").

```python
dataset= pd.read_csv('../data/50_Startups.csv')
```

Del dataset, únicamente nos vamos a quedar con la variable "R&D Spend" como variable predictora y con la variable "Profit" como variable objetivo.

```python
dataset.head()
```

```python
X=dataset.loc[0:,["R&D Spend"]]
y=dataset.loc[0:, ["Profit"]]

print("=="*60)
print(f"shape de X variable independiente {X.shape}")
print(f"shape de y variable Target {y.shape}")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.3)
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scal=StandardScaler()
scal.fit(X_train)
x_train_scaled=scal.transform(X_train)
x_test_scaled=scal.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(x_train_scaled, y_train)
```

```python
#Predicciones con el modelo 
y_pred=regression.predict(x_test_scaled)
print(y_pred.shape)
```

```python
##Genero una grafica visual para ver que a nivel visual que tan bien esta mi módelo
import matplotlib.pyplot as plt

plt.Figure(figsize=(30, 7))
plt.plot(X, y, c="blue", marker="o", alpha=0.9, label="dataset", linestyle="")
plt.plot(X_test, y_pred, c="red", linewidth=1, alpha=0.9)
plt.xlabel("R&D Spend")
plt.ylabel("Profit")
plt.grid(True, c="grey")
plt.legend(loc="upper left")
plt.title("Dataset variacion de profit en función de gastos R&D Spend", fontsize=14)

```
