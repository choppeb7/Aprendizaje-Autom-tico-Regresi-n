# Dividir el dataset en conjunto de entrenamiento y testing

````python
# TODO: Definir las variables X (independientes) y y (dependiente)
# Pista: X debe incluir todas las columnas excepto la variable objetivo
# La variable objetivo es 'house_price_of_unit_area'

X = dataset.loc[0:, ["house_age", "distance_to_the_nearest_station", "number_of_convenience_stores", "latitude", "longitude"]].values  # Completar los √≠ndices de columnas
y = dataset["house_price_of_unit_area"]  # Completar el nombre de la columna objetivo
````

````python
# Verificar las dimensiones de X
print("Dimensiones de X:", X.shape)
print("Dimensiones de y:", y.shape)
````

````python
# TODO: Importar train_test_split desde sklearn
# Pista: from sklearn.model_selection import ???

from sklearn.model_selection import train_test_split
````

````python
# TODO: Dividir los datos en entrenamiento y testing
# Pista: Utilizar test_size = 0.3 y random_state = 0 para reproducibilidad

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=0)
    

print("Tama√±o X_train:", X_train.shape)
print("Tama√±o X_test:", X_test.shape)
print("Tama√±o y_train:", y_train.shape)
print("Tama√±o y_test:", y_test.shape)
````

# Escalado de variables

## ü§î **¬øPor qu√© necesitamos escalar las variables?**
En la regresi√≥n lineal m√∫ltiple, cuando las variables tienen diferentes escalas (a√±os, distancias, coordenadas), es importante normalizarlas para que todas contribuyan de manera equilibrada al modelo.

````python
# TODO: Importar StandardScaler y aplicar escalado
# Pista: from sklearn.preprocessing import ???

from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()

# TODO: ¬øPor qu√© usamos fit_transform en entrenamiento y solo transform en test?
X_train = sc_X.fit_transform(X_train)  # Aprende los par√°metros y transforma
X_test = sc_X.transform(X_test)     # Solo transforma con los par√°metros aprendidos
````

# Ajustar el modelo de Regresi√≥n Lineal M√∫ltiple

````python
# TODO: Importar LinearRegression
# Pista: from sklearn.linear_model import ???

from sklearn.linear_model import LinearRegression
````

````python
# TODO: Crear y entrenar el modelo
# Pista: 1) Crear la instancia del modelo, 2) Ajustar con .fit()

regression = LinearRegression()
regression.fit(X_train, y_train)  # Entrenar el modelo
````

# Predicci√≥n de los resultados en el conjunto de testing

````python
# TODO: Realizar predicciones en el conjunto de test
# Pista: usar .predict() con los datos de test

y_pred = regression.predict(X_test)

print("Predicciones realizadas:", y_pred.shape)
````

## Evaluaci√≥n del modelo: R¬≤ Score (Coeficiente de Determinaci√≥n)

### üìö **Conceptos clave:**
- **R¬≤**: Mide qu√© porcentaje de la variabilidad en Y es explicada por el modelo
- **Rango**: 0 a 1 (1 = modelo perfecto, 0 = modelo in√∫til)
- **Interpretaci√≥n**: R¬≤ = 0.85 ‚Üí el modelo explica el 85% de la varianza

````python
# TODO: Calcular el R¬≤ score
# Pista: Importar r2_score y usarlo con y_true y y_pred

from sklearn.metrics import r2_score

r2 = r2_score(y_true=y_test, y_pred=y_pred)
print(f"R¬≤ Score: {r2:.4f}")
print(f"El modelo explica el {r2*100:.2f}% de la varianza en los datos")
````

## R¬≤ Ajustado: Una m√©trica m√°s robusta

### üéØ **¬øPor qu√© R¬≤ ajustado?**
- R¬≤ simple siempre mejora al a√±adir m√°s variables (aunque no sean √∫tiles)
- R¬≤ ajustado penaliza la adici√≥n de variables irrelevantes
- Es m√°s confiable para modelos con m√∫ltiples variables

````python
# TODO: Implementar funci√≥n para R¬≤ ajustado
# F√≥rmula: R¬≤_adj = 1 - ((1-R¬≤) * (N-1) / (N-p-1))
# donde N = n√∫mero de observaciones, p = n√∫mero de variables independientes

def r2_adjusted(y_true, y_pred, r2, X_test):
    """
    Calcula el R¬≤ ajustado

    Par√°metros:
    - y_true: valores reales
    - y_pred: valores predichos
    - r2: R¬≤ simple
    - X_test: conjunto de test para obtener N y p

    Retorna: R¬≤ ajustado
    """
    N = y_pred.shape[0]  # n√∫mero de observaciones
    p = X_test.shape[1]  # n√∫mero de variables independientes

    # TODO: Completar la f√≥rmula del R¬≤ ajustado
    # Pista: adjusted_r2 = 1 - ((1-r2) * (N-1) / (N-p-1))
    adjusted_r2 = 1 - ((1-r2) * (N-1) / (N-p-1))

    return adjusted_r2
````

````python
# Calcular R¬≤ ajustado
r2_adj = r2_adjusted(y_test, y_pred, r2, X_test)
print(f"R¬≤ Ajustado: {r2_adj:.4f}")
print(f"Diferencia: {(r2 - r2_adj):.4f}")
````
