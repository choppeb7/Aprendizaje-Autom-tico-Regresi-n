# Importar las librerías

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importamos de la librería sklearn, la función train_test_split
from sklearn.model_selection import train_test_split
# Importamos la función StandardScaler para normalizar las variables
from sklearn.preprocessing import StandardScaler
# Importamos la función LinearRegression de la librería sklearn para inicializar el modelo
from sklearn.linear_model import LinearRegression
```

# Importar el dataset

Este conjunto de datos tiene datos recopilados de Nueva York, California y Florida sobre 50 startups, 17 en cada estado. Las variables utilizadas en el conjunto de datos son  el beneficio, Gasto en I+D, Gasto en administración y Gasto en marketing. La variable objetivo a predecir en este caso va a ser el beneficio ("Profit").

```python
df=pd.read_csv('../data/50_Startups.csv')
```

# Parte 1

Del dataset, únicamente nos vamos a quedar con la variable "R&D Spend" como variable predictora y con la variable "Profit" como variable objetivo.

```python
df.head()
```

```python
X=df[["R&D Spend"]].values
y=df["Profit"].values
```

```python
#Verificando que cada uno de los componentes de entrenamiento sea tipo numpy
print("=="*60)
print("Comprobando formato y tipo de dato de los features y el target")
print(f"tipo de dato de feature: {type(X)}, shape de la variable Target {X.shape}")
print(f"tipo de dato de feature: {type(y)}, shape de la variable Target {y.shape}")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
# Antes de entrenar el modelo tenemos que estandarizar las variables independientes
from sklearn.preprocessing import StandardScaler

def escalado(X): 
    scaler=StandardScaler()
    scaler.fit(X)
    X_std=scaler.transform(X)
    return X_std
```

```python
#Entreno el modelo con X_train para eso creo un objeto Linearregression
from sklearn.linear_model import LinearRegression

X_train_std=escalado(X_train)
model=LinearRegression()
model.fit(X_train_std, y_train)
```

```python
coeficientes=model.coef_
print(f"Verificando coeficientes del modelo: {coeficientes[0]:.2f}")
```

```python
#Antes de predecir hay que escalar el dataset de entrenamiento
X_test_std=escalado(X_test)
```

```python
y_predic=model.predict(X_test_std)
```

```python
#Importar métricas de evaluación de error de modelos 
from sklearn.metrics import r2_score, mean_absolute_error

r2=r2_score(y_predic, y_test)
MAE=mean_absolute_error(y_predic, y_test)

print("Resumen evaluación de métricas")
print("=="*60)
print(f"R2 Para el modelo lineal es igual a: {r2:.2f}")
print(f" MAE Para el modelo lineal es igual a: {MAE:.2f}")
```

# Parte 2

Hacemos el mismo ejercicio pero con el gasto en marketing, usando el mismo target

```python
#Ok ahora el mismo proceso solo que con el gasto de marketing

X=df[["Marketing Spend"]].values
y=df["Profit"].values
```

```python
#Evaluacion de tipo de datos y shape
print(f"La variable Feature, Tipo de dato {type(X)}, y el shape de Feature {X.shape}")
print(f"Target, Tipo de dato {type(y)}, y el shape del Target {y.shape}")
```

```python
## Separacion del dataset en entrenamiento y evalaucion
from sklearn.model_selection import train_test_split

X_train2, X_test2, y_train2, y_test2=train_test_split(X,y, test_size=0.3, random_state=43)

```

```python
#Escalado por separado de dataset de entrenamiento y validación
X_train2_std=escalado(X_train2)
X_test2_std=escalado(X_test2)
```

```python
#Entrenamiento del modelo 
from sklearn.linear_model import LinearRegression

model2=LinearRegression()
model2.fit(X_train2_std, y_train2)
```

```python
y_pred2=model2.predict(X_test2_std)
```

```python
#importar librerias para analizar el modelo 
from sklearn.metrics import r2_score, mean_absolute_error

r2=r2_score(y_pred2, y_test2)
mae=mean_absolute_error(y_pred2, y_test2)

print(f"Probando el segundo modelo R2: {r2:.2f}")
print(f"Probando el segundo modelo MAE: {mae:.2f}")
```
